```python
import gdown
import pandas as pd

# 아래 두개의 값 입력 (로컬 컴퓨터 환경에서 실행은 url 비워둘 것)
filename = 'pt_book_fix.csv'  # 예제에서 사용할 filename 입력
url = ''    # 여기에 본인이 가진 링크의 난수 입력

if url != '' :
    gdown.download(f'https://drive.google.com/uc?id={url}', filename, quiet=False)
    dir = ''
else :
    dir = '../_csv/'  # 로컬 컴퓨터 환경에서 사용할 dir (예제는 _csv)

pt_book = pd.read_csv(f'{dir}{filename}', low_memory=False)
pt_book.head()
```

```python
import matplotlib.pyplot as plt

plt.scatter(pt_book['도서권수'], pt_book['대출건수'], alpha=0.1)
plt.show()
```

```python
plt.figure(figsize=(9, 6))  # 너비, 높이 (기본 6, 4) 인치
plt.scatter(pt_book['도서권수'], pt_book['대출건수'], alpha=0.1)
plt.show()
```

여러개의 서브플롯 출력

```python
fig, axs = plt.subplots(2)  # subplots는 2개의 튜플을 리턴 (figure 클래스 객체, axes 클래스 객체 배열)

# 첫번째 그래프
axs[0].scatter(pt_book['도서권수'], pt_book['대출건수'], alpha = 0.1)

# 두번째 그래프
axs[1].hist(pt_book['대출건수'], bins = 100)
axs[1].set_yscale('log')
```

```python
fig, axs = plt.subplots(2, figsize=(6,8))  # figsize 매개변수로 사이즈 조절

# 첫번째 그래프
axs[0].scatter(pt_book['도서권수'], pt_book['대출건수'], alpha = 0.1)   
axs[0].set_title('scatter plot')

# 두번째 그래프
axs[1].hist(pt_book['대출건수'], bins = 100)
axs[1].set_title('histogram')
axs[1].set_yscale('log')
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10,4))  # 두개의 열을 가진 서브플롯

# 첫번째 그래프
axs[0].scatter(pt_book['도서권수'], pt_book['대출건수'], alpha = 0.1)   
axs[0].set_title('scatter plot')
axs[0].set_xlabel('number of books')
axs[0].set_ylabel('borrow count')

# 두번째 그래프
axs[1].hist(pt_book['대출건수'], bins = 100)
axs[1].set_title('histogram')
axs[1].set_yscale('log')
axs[1].set_xlabel('borrow count')
axs[1].set_ylabel('frequency')
```
